```python
def chord_diagram_json_maker(workwhat):
    flare = {}
    for d in pure_data:
        if "livesin" in d:
            if "workwhat" in d:
                if d["workwhat"] == workwhat:
                    birth = d["birth"]
                    lives = d["livesin"]
                    if birth not in flare:
                        flare[birth] = {}
                    if lives not in flare[birth]:
                        flare[birth][lives] = 1
                    else:
                        flare[birth][lives] = flare[birth][lives] + 1
    return flare
                
flare = chord_diagram_json_maker('Sportoló/Edző/Bíró')
print(flare)
```

```python
def chord_diagram_matrix_maker(workwhat):
    flare = chord_diagram_json_maker(workwhat);
    matrix = [[0 for i in range(len(szukitett))] for j in range(len(szukitett))]
    
    for b in flare:
        for l in flare[b]:
            matrix[szukitett.index(b)][szukitett.index(l)] = flare[b][l]
    
    return matrix

m = numpy.array(chord_diagram_matrix_maker('Fodrász/Kozmetikus'))
print(m)
```
